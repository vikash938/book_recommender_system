#### merging ratings and books dataset based on 'ISBN'

```python
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating' : 'number_of_rating'},inplace=True)
```

```python
number_rating
```

```python
### merge based on titie
final_rating = ratings_with_books.merge(number_rating, on = 'title')
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['number_of_rating']>=50]
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['number_of_rating']>=50]
```

```python
final_rating.head()
```

```python
#shape of the dataset
final_rating.shape
```

```python
# DROP DUPLICATES from the dataset
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating.shape
```

```python
## creating a pivot table

book_pivot = final_rating.pivot_table(columns='user_id',index = 'title',values='rating')
```

```python
book_pivot
```

```python
# filling NaN values
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
import scipy
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

#### Model Building

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
```

```python
print(model)
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
# for book's name
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions][i])
```

```python
# all books name
book_pivot.index[393]
```

```python
np.where(book_pivot.index == 'Animal Farm')[0][0]
```

```python
# suggestion based on books name

def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestions)):
        if i == 0:
            print("The suggestions for", book_name, "are :")
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Harry Potter and the Chamber of Secrets (Book 2)')
```
