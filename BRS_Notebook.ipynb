Popularity Based Recommender System

```python
# merging the dataset based on ISBN ID
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.head()
```

```python
# Now groupby based on Book-Title 
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Average Rating
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popularity_df = popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num_ratings','avg_rating']]
```

Collabrative Filtering Based Recommender System

- Books - book name
- Ratings - user_id, book_id, rating
- user in columns
- books in rows

```python
ratings_with_name
```

```python
# we want only that users who gave minimum 200 votes
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200

educated_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
```

```python
# filtering based on book title
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

Every Book become a vector of 810 dimensional space

Here for book1 810 vectors
and also book2 has 810 vectors
and so on 
so we calculate euclidian distance between every vectors respect to every vectors

```python
pt.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# cosine_similarity(pt).shape
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores[0]
```

Here 1st vector book 1984 kaa khud ke sath or baaki sabhi books ke sath similarity score bta rha h.

```python
# name of the movie   pt.index[47]


# def recommend(book_name):
#   # index_fetch
#   index = np.where(pt.index==book_name)[0][0]
#   # distances = similarity_scores[index]
#   similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:10]
#   for i in similar_items:
#     print(pt.index[i[0]])

def recommend(book_name):
  # index_fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:10]
  
  
  data = []
  for i in similar_items:
    item = []
    temp_df = books[books['Book-Title'] == pt.index[i[0]]]
    item.extend(temp_df.drop_duplicates('Book-Title')['Book-Title'])
    item.extend(temp_df.drop_duplicates('Book-Title')['Book-Author'])
    item.extend(temp_df.drop_duplicates('Book-Title')['Image-URL-M'])
    item.extend(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'])


    data.append(item)

  return data
```

```python
recommend('Message in a Bottle')
```

```python
recommend('1984')
```

```python
recommend('The Notebook')
```

```python
recommend('The Da Vinci Code')
```
